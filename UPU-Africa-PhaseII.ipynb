> Part I Evaluation of Performance Quality 

>> Calculation of Performance Quality for each Country 
>>1. Perform a formula for calculating the total of each one of the performance indicators per country.
>>2. Add a new column for the total of each one of the performance indicators per country.
>>3. Create a new Dataframe for the total of each one the indicators and sub-indicators for the entire region.

>>> Prepare a new data frame with
>>>1. all the indicators columns 
>>>2. create new columns for the values of the calculated overall performance indicators
>>>3. calculate the percentage of the values of the overall indicators

```python
# Select the columns needed for the evaluation of the performance quality
eval_data = africa_data[['Country', 1.1, 1.2, 1.3, 2.1, 2.2, 2.3, 2.4, 3.1, 3.2, 3.3, 4.1, 5.1]]
eval_data
```

```python
### 1. Replace the values of the indicators to integers (1-5)
replacer = {'Strongly Disagree': 1, 'Disagree': 2,
            'Neutral': 3,
            'Agree': 4, 'Strongly Agree': 5, 'Pas du tout d’accord': 1, 'Pas d’accord': 2,
            'Neutre': 3,
            'D’accord': 4, 'Tout à fait d’accord': 5}

# Replace Values in the evaluation Columns
eval_data = eval_data.replace(replacer)
eval_data.fillna(-1, inplace=True)
eval_data.drop(23, inplace=True)
eval_data
```

```python
# 2. Calculate the overall performance quality for each indicator per country
eval_data['Ben_Indicator'] = (eval_data[1.1] + eval_data[1.2] + eval_data[1.3])/15*100
eval_data['Cust_Indicator'] = (eval_data[2.1] + eval_data[2.2] + eval_data[2.3] + eval_data[2.4])/20*100
eval_data['Open_Indicator'] = (eval_data[3.1] + eval_data[3.2] + eval_data[3.3])/15*100
eval_data['Compens_Indicator'] = (eval_data[4.1])/5*100
eval_data['Coverage'] = eval_data[5.1].str.replace("%","")
eval_data['Coverage'] = eval_data['Coverage'].astype(int)
eval_data['Overall'] = (eval_data['Ben_Indicator'] + eval_data['Cust_Indicator'] + eval_data['Open_Indicator'] + eval_data['Compens_Indicator'] + eval_data['Coverage'])//5
eval_data
```

```python
# Data Preparation
qual_ind_country = indicators_data = eval_data[['Country', "Ben_Indicator", "Cust_Indicator", "Open_Indicator", "Compens_Indicator", "Coverage"]]
qual_ind_country.set_index('Country', inplace=True)
# Rename the column headers
qual_ind_country.rename(columns = {'Ben_Indicator':'Beneficiaries', 'Cust_Indicator':'Customer Complaints', 'Open_Indicator': 'Opening Hours', 'Compens_Indicator': 'Compensation', 'Coverage': 'Network Coverage', 'Overall':'Overall Evaluation'}, inplace = True)

qual_ind_country
```

```python
# Chart Preparation
country_names =qual_ind_country.columns

# Initialize an empty dictionary
distribution = {}

# Iterate through the rows of the DataFrame cust_country 
# to set values of the dictionary distribution
for index, row in qual_ind_country.iterrows():
    # Assuming 'Country' column as the key and rest of the row as the value
    key = qual_ind_country.loc[index].name # Country 
    value = qual_ind_country.loc[index].tolist() # values
    distribution[key] = value  # add a new item to the dictionary

# Formatting and Styling of the Pie Chart
colors_list = ['darkcyan', 'maroon', 'coral', 'olive', 'green']
text_font = {'family':'Arial','color':'black','size':10, 'weight':'normal'}
#title_font = {'family':'fantasy','color':'black','size':20, 'weight':'normal'}
legend_font = {'family':'Times New Roman','size':10, 'weight':'normal'}

def draw(distribution, country_names):
    """
   
    """
    labels = list(distribution.keys()) # Retrieve the Country as a list
    data = np.array(list(distribution.values())) # Retrieve the counter of a customer type as an array of lists
    data_cum = data.cumsum(axis=1)
    cust_colors = plt.colormaps['seismic'](
        np.linspace(0.15, 0.85, data.shape[1]))

    fig, ax = plt.subplots(figsize=(9.2, 5))
    # Plotting + formatting the overall plot
    # Grid formatting
    
    #ax.yaxis.grid(True, linestyle='--', which='major', color='grey', alpha=.25)
    #ax.xaxis.grid(True, linestyle='--', which='major', color='grey', alpha=.25)
    ax.spines['bottom'].set_color('black') # Color the bottom axis lines
    ax.spines['left'].set_color('black') # Color the left axis lines

    ax.invert_yaxis()
    ax.xaxis.set_visible(False)
    ax.set_xlim(0, np.sum(data, axis=1).max())

    for i, (colname, color) in enumerate(zip(country_names, colors_list)):
        widths = data[:, i]
        starts = data_cum[:, i] - widths
        rects = ax.barh(labels, widths, left=starts, height=0.5,
                        label=colname, color=color)

        # Generate colors for the text inside the bars
        #r, g, b, _ = color
        #text_color = 'white' if r * g * b < 0.5 else 'darkgrey'
        # display and format label values in the bars
        #ax.bar_label(rects, label_type='center', font='Arial', color=text_color)
    
    ax.set_yticklabels(labels, rotation=0, fontsize=10, fontweight='normal', 
                  fontname='Times New Roman', color='black')
    #ax.set_xticklabels(data, rotation=0, fontsize=10, fontweight='normal', 
    #             fontname='Times New Roman', color='black')
    
    ax.set_xlabel('Performance Indicators', weight='bold', fontsize = 10, color = 'black', font='Cambria') # add to x-label to the plot)
    ax.set_ylabel('Country', weight='bold', fontsize = 12, color = 'black', font='Cambria') # add to y-label to the plot
    ax.legend(ncols=len(country_names), bbox_to_anchor=(-0.1, 1),
              loc='lower left', prop=legend_font)
    

    return fig, ax


draw(distribution, country_names)
plt.show()

############################ Done ############################
```

```python
# Data Preparation
indicators_data = eval_data[[1.1, 1.2, 1.3, 2.1, 2.2, 2.3, 2.4, 3.1, 3.2, 3.3, 4.1, "Ben_Indicator", "Cust_Indicator", "Open_Indicator", "Compens_Indicator", "Coverage", "Overall"]]
#Calculate the average of each one of the indicators and sub-indicators for all countries.
sum_ind = indicators_data.mean()
# Re-scale the average of each one of the sub-indicators for all countries to become out of 100%.
sum_ind[1.1] = sum_ind[1.1]/5*100
sum_ind[1.2] = sum_ind[1.2]/5*100
sum_ind[1.3] = sum_ind[1.3]/5*100
sum_ind[2.1] = sum_ind[2.1]/5*100
sum_ind[2.2] = sum_ind[2.2]/5*100
sum_ind[2.3] = sum_ind[2.3]/5*100
sum_ind[2.4] = sum_ind[2.4]/5*100
sum_ind[3.1] = sum_ind[3.1]/5*100
sum_ind[3.2] = sum_ind[3.2]/5*100
sum_ind[3.3] = sum_ind[3.3]/5*100
sum_ind[4.1] = sum_ind[4.1]/5*100

# Create a dataframe for the table and chart

# initialize data of lists.
data = {'Cat': ['Beneficiaries', 'Beneficiaries', 'Beneficiaries', 
                'Customer Complaints', 'Customer Complaints', 'Customer Complaints', 'Customer Complaints', 
                'Opening Hours', 'Opening Hours', 'Opening Hours',
                'Compensation', 'Network Coverage'],
        'SubCat': ['Send/recieve at home', 'Send/recieve all week', 'Send/recieve once a day', 
                'Transparent procedure', 'Complaints for Improvement ', 'Eliciting needs', 'Needs for improvement ', 
                'Offices for 5 days', 'Franchise for 5 days', 'Informing changing hours',
                'Compensation', 'Network Coverage'],
        'Eval' : [sum_ind[1.1], sum_ind[1.2], sum_ind[1.3],
                  sum_ind[2.1], sum_ind[2.2], sum_ind[2.3], sum_ind[2.4],
                  sum_ind[3.1], sum_ind[3.2], sum_ind[3.3],
                  sum_ind[4.1],
                  sum_ind["Coverage"]
                 ]
       }

# Create DataFrame
indicators_table = pd.DataFrame(data)

# Rename the column names
indicators_table.rename(columns = {'Cat':'Indicators', 'SubCat':'Short Statement', 'Eval': 'Evaluation%'}, inplace = True)
#transform the table into pivot table
eval_pivot = indicators_table.pivot_table(index=['Indicators','Short Statement'],
                        values=['Evaluation%'],
                        aggfunc='sum')

##########################################
# Saving the table into an excel sheet
with pd.ExcelWriter('Phase2Tables.xlsx', mode='a') as writer:
    eval_pivot.to_excel(writer, sheet_name='Evaluation per Indicator')  

eval_pivot

################# Done #######################
```
